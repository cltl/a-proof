## Getting the cohen kappa score for all annotations and per annotation label

We first get the list of annotations from the data frames and obtain the unique set of labels

````python
labels1 = util.get_token_anno(df1)
labels2 = util.get_token_anno(df2)
print(len(labels1))
print(len(labels2))
````

````python
### Unique set of labels
annotation_labels = set(labels1+labels2)
print(annotation_labels)
````

### Kappa score across all labels (including "_"

````python
#kappa=cohen_kappa_score(labels1, labels2, weights='quadratic')
kappa=cohen_kappa_score(labels1, labels2)
print('Kappa:',kappa)
````

### Kappa score per label

To get the kappa score per label, we iterate over the set and check for each row if one of the two data frames has this label.
If so, both values are copied to the value lists for each frame. We use these value list for the kapa score for that label.

````python
def get_kappa_for_label (df1, df2, label):
    values1=[]
    values2=[]
    for index, token in enumerate(df1['sen_id-tok_id']):
        label1=df1['annotation'].iloc[index]
        label2=df2['annotation'].iloc[index]
        if df1['sen_id-tok_id'].iloc[index]!=df2['sen_id-tok_id'].iloc[index]:
            ### we check if the token ids are actually the same across the data frame rows.
            ### if not we print a warning
            print("Warning! Token identifiers of rows do not match:")
            print(df1['sen_id-tok_id'].iloc[index], df2['sen_id-tok_id'].iloc[index])
        else:
            if (label1==label):
                values1.append(label1)
                values2.append(label2)
            elif label2==label:
                values1.append(label1)
                values2.append(label2)
    kappa=cohen_kappa_score(values1, values2)
    return kappa
````

````python
kappas={}
for label in annotation_labels:
    kappa=get_kappa_for_label(df1, df2, label)
    kappas[label]=kappa
````
